```python
import pandas as pd
import plotly.express as px

# Assuming you already have the 'df' DataFrame

# Sum the DALYs for each disorder
df['Total DALYs from Depressive Disorders'] = df['DALYs from depressive disorders per 100,000 people in, both sexes aged age-standardized'].sum()
df['Total DALYs from Schizophrenia'] = df['DALYs from schizophrenia per 100,000 people in, both sexes aged age-standardized'].sum()
df['Total DALYs from Bipolar Disorder'] = df['DALYs from bipolar disorder per 100,000 people in, both sexes aged age-standardized'].sum()
df['Total DALYs from Eating Disorders'] = df['DALYs from eating disorders per 100,000 people in, both sexes aged age-standardized'].sum()
df['Total DALYs from Anxiety Disorders'] = df['DALYs from anxiety disorders per 100,000 people in, both sexes aged age-standardized'].sum()

# Create a new DataFrame for total DALYs for each disorder
total_daly_df = pd.DataFrame({
    'Disorder': ['Depressive Disorders', 'Schizophrenia', 'Bipolar Disorder', 'Eating Disorders', 'Anxiety Disorders'],
    'Total DALYs': [
        df['Total DALYs from Depressive Disorders'].values[0],
        df['Total DALYs from Schizophrenia'].values[0],
        df['Total DALYs from Bipolar Disorder'].values[0],
        df['Total DALYs from Eating Disorders'].values[0],
        df['Total DALYs from Anxiety Disorders'].values[0]
    ]
})

# Bar plot for the total DALYs for each disorder
fig = px.bar(total_daly_df, x='Disorder', y='Total DALYs',
             title='Total DALYs for Mental Health Disorders',
             labels={'Total DALYs': 'Total DALYs per 100,000 people', 'Disorder': 'Mental Health Disorder'},
             color='Disorder')



fig.show()
```

```python
fig.write_html('viz1.html')
```

```python
# Sum the DALYs for each disorder
df['Total DALYs'] = df[['DALYs from depressive disorders per 100,000 people in, both sexes aged age-standardized',
                        'DALYs from schizophrenia per 100,000 people in, both sexes aged age-standardized',
                        'DALYs from bipolar disorder per 100,000 people in, both sexes aged age-standardized',
                        'DALYs from eating disorders per 100,000 people in, both sexes aged age-standardized',
                        'DALYs from anxiety disorders per 100,000 people in, both sexes aged age-standardized']].sum(axis=1)

# Create a Choropleth map for total DALYs
fig = px.choropleth(df, 
                    locations='Entity',
                    locationmode='country names',
                    color='Total DALYs',
                    hover_name='Entity',
                    title='Total DALYs for Mental Health Disorders Across Countries',
                    labels={'Total DALYs': 'Total DALYs per 100,000 people', 'Entity': 'Country'},
                    color_continuous_scale='Viridis')

fig.show()
```

```python
import plotly.express as px

# Select relevant columns for counting
columns_to_count = ['DALYs from depressive disorders per 100,000 people in, both sexes aged age-standardized',
                    'DALYs from schizophrenia per 100,000 people in, both sexes aged age-standardized',
                    'DALYs from bipolar disorder per 100,000 people in, both sexes aged age-standardized',
                    'DALYs from eating disorders per 100,000 people in, both sexes aged age-standardized',
                    'DALYs from anxiety disorders per 100,000 people in, both sexes aged age-standardized']

# Create a new column for the total count across all disorders
df['Total DALYs'] = df[columns_to_count].sum(axis=1)

# Create and save choropleth maps for each disorder
for column in columns_to_count:
    fig = px.choropleth(df, 
                        locations='Entity',
                        locationmode='country names',
                        color=column,
                        hover_name='Entity',
                        title=f'{column} Across Countries',
                        labels={'Entity': 'Country', column: f'{column} per 100,000 people'},
                        color_continuous_scale='Viridis')
    fig.write_html(f'{column.replace(" ", "_").lower()}_map.html')

```
